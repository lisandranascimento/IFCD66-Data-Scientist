**Tarea:**

1. Basándote en el código anterior, escribe una función que genere números aleatorios uniformemente distribuidos. Hay varios requisitos para tu función:
    * Debe aceptar 3 parámetros: 
        * `bottom` - el límite inferior de los números generados
        * `ceiling` - el límite superior de los números generados
        * `count` - cuántos números generar
    * Debe devolver una matriz de números aleatorios distribuidos uniformemente

2. Llama a tu función con 2 conjuntos de parámetros a continuación:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

3. Traza las distribuciones uniformes generadas anteriormente utilizando histogramas, donde el eje x es el valor y el eje y es el recuento. Que el número de intervalos del histograma sea 10.

El resultado debe ser como el siguiente:



```python
# your code here
def generate_uniform_random_numbers(bottom, ceiling, count):
    """
    Genera números aleatorios uniformemente distribuidos entre bottom y ceiling.
    
    Parámetros:
    bottom (float): Límite inferior de los números generados.
    ceiling (float): Límite superior de los números generados.
    count (int): Cantidad de números a generar.
    
    Retorna:
    numpy.ndarray: Matriz de números aleatorios distribuidos uniformemente.
    """
    return np.random.uniform(low=bottom, high=ceiling, size=count)

# Llamada a la función con el primer conjunto de parámetros
uniform_numbers_1 = generate_uniform_random_numbers(bottom=10, ceiling=15, count=100)

# Llamada a la función con el segundo conjunto de parámetros
uniform_numbers_2 = generate_uniform_random_numbers(bottom=10, ceiling=60, count=1000)

# Configuración de los gráficos
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Histograma para el primer conjunto de números
axs[0].hist(uniform_numbers_1, bins=10, edgecolor='black')
axs[0].set_ylim(0, 130)

# Histograma para el segundo conjunto de números
axs[1].hist(uniform_numbers_2, bins=10, edgecolor='black')
axs[1].set_ylim(0, 130)

# Ajuste de los gráficos
plt.tight_layout()
plt.show()

```

How are the two distributions different?

Las diferencias principales se deben al rango de valores y la cantidad de números generados. La primera distribución tiene un rango más estrecho y menos números, lo que resulta en un histograma con barras más agrupadas y variabilidad en altura. La segunda distribución tiene un rango más amplio y más números, resultando en un histograma más uniforme y disperso.

## Distribución Normal

1. De la misma manera en el reto de Distribución Uniforme, escriba una función que genere números aleatorios distribuidos normalmente.
1. Genera 1.000 números distribuidos normalmente con media de 10 y desviación típica de 1
1. Genera 1.000 números distribuidos normalmente con una media de 10 y una desviación típica de 50
2. Representa gráficamente las distribuciones de los datos generados.

Resultado esperado:



```python
from scipy.stats import norm 
```

```python
# your code here
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def generate_normal_random_numbers(mean, std_dev, count):
    """
    Genera números aleatorios distribuidos normalmente con una media y desviación estándar dadas.
    
    Parámetros:
    mean (float): Media de la distribución.
    std_dev (float): Desviación estándar de la distribución.
    count (int): Cantidad de números a generar.
    
    Retorna:
    numpy.ndarray: Matriz de números aleatorios distribuidos normalmente.
    """
    return norm.rvs(loc=mean, scale=std_dev, size=count)

# Generar 1,000 números distribuidos normalmente con media de 10 y desviación estándar de 1
normal_numbers_1 = generate_normal_random_numbers(mean=10, std_dev=1, count=1000)

# Generar 1,000 números distribuidos normalmente con media de 10 y desviación estándar de 50
normal_numbers_2 = generate_normal_random_numbers(mean=10, std_dev=50, count=1000)

# Configuración de los gráficos
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Histograma para el primer conjunto de números
axs[0].hist(normal_numbers_1, bins=30, edgecolor='black')

# Histograma para el segundo conjunto de números
axs[1].hist(normal_numbers_2, bins=30, edgecolor='black')

# Ajuste de los gráficos
plt.tight_layout()
plt.show()
```

How are the two distributions different?

La primera distribución con una desviación estándar de 1 es mucho más concentrada alrededor de la media de 10, formando una campana estrecha y alta en el histograma. La segunda distribución con una desviación estándar de 50 es mucho más dispersa, abarcando un rango amplio de valores, lo que resulta en un histograma más ancho y bajo. Estas diferencias reflejan cómo la desviación estándar afecta la dispersión de los datos en una distribución normal.

¿Cuál o cuáles de las variables tienen una distribución casi normal? ¿Cómo lo sabes?

```python
# your code here
# Lista de variables que queremos inspeccionar
variables = ['Fuel Barrels/Year', 'Combined MPG', 'CO2 Emission Grams/Mile']  

# Configuración de los gráficos
fig, axs = plt.subplots(len(variables), 1, figsize=(4, 2 * len(variables)))

for i, var in enumerate(variables):
    sns.histplot(df[var], bins=30, kde=True, ax=axs[i])
    axs[i].set_title(f'Histograma de {var}')
    axs[i].set_xlabel(var)
    axs[i].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

```python
# your code here
import scipy.stats as stats

# Función para realizar el test de Shapiro-Wilk y retornar los resultados
def test_normality(data, alpha=0.05):
    stat, p = stats.shapiro(data)
    print(f'Estadístico de Shapiro-Wilk: {stat}, p-valor: {p}')
    if p > alpha:
        print('La muestra parece provenir de una distribución normal (no se rechaza H0)')
    else:
        print('La muestra no parece provenir de una distribución normal (se rechaza H0)')

# Realizar el test de Shapiro-Wilk para cada variable
for var in variables:
    print(f'\nResultados del test de Shapiro-Wilk para {var}:')
    test_normality(df[var].dropna())
```

Teniendo en cuenta que:

a) Histograma y Curva KDE: Si la forma del histograma y la curva KDE es aproximadamente una campana simétrica, la variable podría ser normalmente distribuida.

b) Prueba de Shapiro-Wilk: Si el p-valor es mayor que 0.05, no rechazamos la hipótesis nula y consideramos que la distribución de la variable es aproximadamente normal.

Podemos decir que ninguno de ellos está normalmente ditribuido. 

## Distribución Exponencial

1. Utilizando `numpy.random.exponential`, crea una función que devuelva una lista de números distribuidos exponencialmente con la media de 10. 

1. Utiliza la función para generar dos secuencias de números de tamaño 10 y 100.

1. 2. Traza las distribuciones como histogramas con el número de bins como 100.

El resultado debe ser como el siguiente:



```python
def generate_exponential_random_numbers(mean, size):
    """
    Genera números aleatorios distribuidos exponencialmente con una media dada.
    
    Parámetros:
    mean (float): Media de la distribución exponencial.
    size (int): Cantidad de números a generar.
    
    Retorna:
    numpy.ndarray: Matriz de números aleatorios distribuidos exponencialmente.
    """
    return np.random.exponential(scale=mean, size=size)

# Generar dos secuencias de números distribuidos exponencialmente
exp_numbers_10 = generate_exponential_random_numbers(mean=10, size=10)
exp_numbers_100 = generate_exponential_random_numbers(mean=10, size=100)

# Configuración de los gráficos
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Histograma para la primera secuencia (tamaño 10)
axs[0].hist(exp_numbers_10, bins=100, edgecolor='black', density=True)
axs[0].set_title('Distribución Exponencial (size=10, mean=10)')
axs[0].set_xlabel('Valor')
axs[0].set_ylabel('Densidad')

# Histograma para la segunda secuencia (tamaño 100)
axs[1].hist(exp_numbers_100, bins=100, edgecolor='black', density=True)
axs[1].set_title('Distribución Exponencial (size=100, mean=10)')
axs[1].set_xlabel('Valor')
axs[1].set_ylabel('Densidad')

# Ajuste de los gráficos
plt.tight_layout()
plt.show()
```

## Distribución Exponencial de datos reales

Supongamos que el tiempo que uno pasa en un banco se distribuye exponencialmente con media de 10 minutos (es decir, λ = 1/10). ¿Cuál es la probabilidad de que un cliente pase menos de quince minutos en el banco? 

Escribe un código en python para resolver este problema

```python
# your code here
import numpy as np

# Definir los parámetros
media = 10  # Media en minutos
lambd = 1 / media  # Parámetro lambda para la distribución exponencial

# Calcular la probabilidad de que X < 15 minutos
x = 15  # Tiempo en minutos

probabilidad = 1 - np.exp(-lambd * x)
```

¿Cuál es la probabilidad de que el cliente pase más de 15 minutos?

```python
# your code here
print(f"La probabilidad de que un cliente pase menos de 15 minutos en el banco es: {probabilidad:.4f}")
```
